```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd
import numpy as np

def leaflet_plot_stations(hashid):

    df = pd.read_csv('BinSize_d400.csv')

    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()

leaflet_plot_stations('fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89')
```

```python
data = pd.read_csv('data/C2A2_data/BinnedCsvs_d400/fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89.csv')
data
```

```python
data['Data_Value'] = data['Data_Value']*0.1
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data['Year'] = data['Date'].dt.year
data['Month_day'] = data['Date'].dt.strftime('%m-%d')
data
```

```python
data = data[data['Month_day'] != '02-29']
data
```

```python
max_temp = data[(data.Year>=2005) & (data.Year<2015) & (data['Element'] == 'TMAX')].groupby(['Month_day'])['Data_Value'].max()
min_temp = data[(data.Year>=2005) & (data.Year<2015) & (data['Element'] == 'TMIN')].groupby(['Month_day'])['Data_Value'].min()
```

```python
data = data.merge(max_temp.reset_index(drop=False).rename(columns={'Data_Value':'Max_temp'}), on='Month_day', how='left')
data = data.merge(min_temp.reset_index(drop=False).rename(columns={'Data_Value':'Min_temp'}), on='Month_day', how='left')
```

```python
data
```

```python
Highdata =  data[(data.Year==2015) & (data.Data_Value>data.Max_temp)]
Highdata
Lowdata = data[(data.Year==2015) & (data.Data_Value<data.Min_temp)]
Lowdata
```

```python
date_index = np.arange('2005-01-01','2016-01-01', dtype='datetime64[D]')
plt.figure()
plt.plot(date_index,max_temp,color='lightcoral', linewidth=1) 
plt.plot(date_index,min_temp,color='skyblue', linewidth=1)

plt.scatter(Highdata.Date.values,Highdata.Data_Value.values, color='red', s=8)
plt.scatter(Lowdata.Date.values, Lowdata.Data_Value.values, color='blue', s=8)


ax = plt.gca()
ax.axis(['2015-01-01', '2015-12-31', -50,50])

plt.xlabel('Date')
plt.ylabel('° Celsius')
plt.title('Temperature in Ann Arbour Michigan (2005-2015)')

plt.legend(['Record High (2005-2014)', 'Record Low (2005-2014)', 'Record breaking high in 2015', 'Record breaking low in 2015'], 
           loc=0, frameon=False)
ax.fill_between(date_index, max_temp, min_temp, facecolor='grey', alpha=0.25)

x = plt.gca().xaxis

# rotate the tick labels for the x axis
for item in x.get_ticklabels():
    item.set_rotation(45)
```
